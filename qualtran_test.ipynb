```python
from qualtran.bloqs.cryptography.rsa import ModExp
from qualtran.drawing import show_bloq

N = 13*17 # integer to factor
g = 8

mod_exp = ModExp(base=g, mod=N, exp_bitsize=32, x_bitsize=32)
show_bloq(mod_exp)
```

```python
from qualtran.resource_counting import get_cost_value, QubitCount, GateCounts, QECGatesCost
from qualtran.bloqs.basic_gates import Toffoli
from qualtran.resource_counting.generalizers import ignore_split_join, ignore_cliffords, ignore_alloc_free, generalize_cvs
```

```python
gc = get_cost_value(mod_exp, QECGatesCost(),generalizer=(ignore_split_join, ignore_alloc_free, generalize_cvs))
nb_qubits = get_cost_value(mod_exp, QubitCount())
gc
```

```python
from qualtran.resource_counting import get_cost_value, QubitCount, BloqCount
from qualtran.bloqs.basic_gates import CNOT, Toffoli

n_qubits = get_cost_value(mod_exp, QubitCount())
print("Logical qubits:", n_qubits)

```
